```python
# 
import pandas as pd
text_ana = pd.read_csv('text_analysis_twitter.csv')
text_ana = text_ana.drop(['Unnamed: 0'], axis=1)
```

```python
text_ana.head()
```

```python
#
import re

def preprocess_text(text):    
    # 
    text = text.lower()   
    # URL
    text = re.sub(r'http\S+', '', text)    
    return text
```

```python
#
from sklearn.feature_extraction.text import CountVectorizer

# CountVectorizer
vectorizer = CountVectorizer()

# 

text_ana['words'] = text_ana['text'].apply(preprocess_text)


X = vectorizer.fit_transform(text_ana['words'])

# 
# print(':', X.shape)

# 
# print(':', vectorizer.get_feature_names_out())
```

```python
# vectorizer.get_feature_names_out()
vectorizer.get_feature_names_out()
```

```python
#
# 
feature_words = vectorizer.get_feature_names_out()

# 
word_freq = dict(zip(feature_words, X.sum(axis=0).A1))
```

```python
# 
sorted (word_freq.items(), key=lambda x: x[1], reverse=True)
```

```python
# 
import pandas as pd
text_ana = pd.read_csv('text_analysis_weibo.csv', index_col = 0)
# text_ana = text_ana.drop(['Unnamed: 0'], axis=1)
text_ana
```

```python
import jieba

def word_segmentation(text):
    words = jieba.lcut(text)
    return ' '.join(words)

text_ana['/'] = text_ana['/'].iloc[:100]\
                                .astype(str).apply(word_segmentation)
text_ana['/'].head()
```

```python
text = list(text_ana['/'].dropna())
text
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# CountVectorizer
vectorizer = CountVectorizer()

# 
X = vectorizer.fit_transform(text)

# 
print(':', X.shape)
```

```python
# 
print(':', vectorizer.get_feature_names_out())
```

```python
list(X.toarray())
```

```python
# 
feature_words = vectorizer.get_feature_names_out()

# 
word_freq = dict(zip(feature_words, X.sum(axis=0).A1))
```
